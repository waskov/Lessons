```python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
import datetime
count = 0
link = 'https://author.today/work/genre/all?eg=-&fnd=false&page='
books_rate = pd.DataFrame(columns=[ 'CurrentDate','DateLastUpdate','DateOpen', 'NumRate','NameBook','NameAuthor','Volume','Likes','Coments','Looks','Hours','LinkBook','Price', 'Gift'])
for page in range(1,2):
    link_page = link + str(page)
    books = rq.get(link_page, verify = False)
    soup = bs(books.text,'html.parser')
    list_books = soup.find_all('div', class_ = 'book-row')
print('Done')
```
