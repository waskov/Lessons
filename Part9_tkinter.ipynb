# Создадим калькулятор

Калькулятор который умеет складывать, умножать, вычетать, делить два числа. А в случае, если введены не числа, выдает ошибку. Имеет кнопку очистки результатов. 

```python
import tkinter as tk
window = tk.Tk()
window.geometry('600x400+200+100')
window.title("Calculator")

 
e1 = tk.Entry(width=12)
e2 = tk.Entry(width=12)
e1.place(x=100, y=30)
e2.place(x=200, y=30)



label_created_by = tk.Label()
label_created_by.place(x=500, y=350)
label_created_by.config(text= 'created by: VB',
                   font="Arial 10")

label = tk.Label()
label.place(x=300, y=10)

label_symbol = tk.Label()
label_symbol.place(x=175, y=10)

label_error = tk.Label()
label_error.place(x=300, y=10)

font_size="Arial 28"

def is_number(a, b):
    try:
        float(a)
        float(b)
        return True
    except ValueError:
        return False

def minus():
    x=e1.get()
    y=e2.get()
    if is_number(x,y) == True:
        result=float(x)-float(y)
        error = label_error.config(text= '',
                   font=font_size)
        answer = label.config(text= '='+str(result),
                   font=font_size)
        symbol = label_symbol.config(text= '-',
                   font=font_size)
        
    else:
        clear()
        error = label_error.config(text= 'error type',
                   font=font_size)
    
    
def plus():
    x=e1.get()
    y=e2.get()
    if is_number(x,y) == True:
        result=float(x)+float(y)
        error = label_error.config(text= '',
                   font=font_size)
        answer = label.config(text= '='+str(result),
                   font=font_size)
        symbol = label_symbol.config(text= '+',
                   font=font_size)
        
    else:
        clear()
        error = label_error.config(text= 'error type',
                   font=font_size)
def multiply():
    x=e1.get()
    y=e2.get()
    if is_number(x,y) == True:
        result=float(x)*float(y)
        error = label_error.config(text= '',
                   font=font_size)
        answer = label.config(text= '='+str(result),
                   font=font_size)
        symbol = label_symbol.config(text= '*',
                   font=font_size)
        
    else:
        clear()
        error = label_error.config(text= 'error type',
                   font=font_size)
        
def division():
    x=e1.get()
    y=e2.get()
    if is_number(x,y) == True:
        result=float(x)/float(y)
        error = label_error.config(text= '',
                   font=font_size)
        answer = label.config(text= '='+str(result),
                   font=font_size)
        symbol = label_symbol.config(text= '/',
                   font=font_size)
        
    else:
        clear()
        error = label_error.config(text= 'error type',
                   font=font_size)
    
def clear():
    e1.delete(0, tk.END)
    e2.delete(0, tk.END)
    answer = label.config(text='')
    symbol = label_symbol.config(text= '')
    error = label_error.config(text= '',
                   font=font_size)
    
    
button_plus = tk.Button(
    text="plus",
    width=10,
    height=5,
    bg="blue",
    fg="yellow")

button_minus = tk.Button(
    text="minus",
    width=10,
    height=5,
    bg="blue",
    fg="yellow")

button_multiply = tk.Button(
    text="multiply",
    width=10,
    height=5,
    bg="blue",
    fg="yellow")

button_division = tk.Button(
    text="division",
    width=10,
    height=5,
    bg="blue",
    fg="yellow")

button_clear = tk.Button(
    text="clear",
    width=10,
    height=5,
    bg="blue",
    fg="yellow")




button_division.place(x=100, y=80)
button_multiply.place(x=200, y=80)
button_minus.place(x=100, y=180)
button_plus.place(x=200, y=180)
button_clear.place(x=300, y=80)


button_multiply['command']=multiply
button_plus['command']=plus
button_division['command']=division
button_minus['command']=minus
button_clear['command']=clear


window.mainloop()
```
